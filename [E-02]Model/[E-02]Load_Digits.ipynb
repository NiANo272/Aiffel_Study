```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd



#Feature Data 지정하기
digits = load_digits()
digits_data = digits.data

#Label Data 지정하기
digits_label = digits.target

#Target Names 출력해 보기
print(digits.target_names)

#데이터 Describe 해 보기
print(digits.DESCR)

#데이터 프레임 만들기
digits_df = pd.DataFrame(data = digits_data, columns = digits.feature_names)
digits_df

#X_train, X_test, y_train, y_test를 생성
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))

#Decision Tree 사용해 보기
decision_tree = DecisionTreeClassifier(random_state=32)

decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Decision Tree의 결과 값 : ", accuracy)

#Random Forest 사용해 보기
random_forest = RandomForestClassifier(random_state=32)

random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Random Forest의 결과 값 : ", accuracy)

#SVM 사용해 보기
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SVM의 결과 값 : ", accuracy)

#SGD Classifier 사용해 보기
sgd_model = SGDClassifier()

sgd_model.fit(X_train, y_train)

y_pred = sgd_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SGD의 결과 값 : ", accuracy)

#Logistic Regression 사용해 보기
logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression의 결과 값 : ", accuracy)

#학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요?모델의 성능을 평가하는 지표로는 무엇이 좋을까요?
#sklearn.metrics 에서 제공하는 평가지표 중 적절한 것을 선택해 보세요. 선택하신 이유도 설명해 주세요.
print(classification_report(y_test, y_pred))

#가장 높은 결과값을 도출하는 것은 SVM이 0.9888888888888889로 가장 높은 값이 나왔다. 하지만 다른 모델들 또한 높은 지표를 보여주기에
#현재 문제를 해결하기에 준수한 성능이라고 볼 수 있다.
#Precision의 경우 대부분 높은 정확도가 나왔지만 5에 대해 낮은 정확도를 보이며, Recall의 경우 또한 8에 대한 정확도만 낮은 결과를 보여준다. 
#따라서 두 평가 지표 모두 비슷하지만 다른분야에서 특이하게 낮은 결과를 가지고 있기에 두 지표를 조화평균한 f1-score가 가장 적절하다고 생각한다.
```
