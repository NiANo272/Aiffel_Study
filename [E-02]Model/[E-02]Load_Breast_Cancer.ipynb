```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd

#Feature Data 지정하기
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data

#Label Data 지정하기
breast_cancer_label = breast_cancer.target

#Target Names 출력해 보기
print(breast_cancer.target_names)

#데이터 Describe 해 보기
print(breast_cancer.DESCR)

#데이터 프레임 만들기
breast_cancer_df = pd.DataFrame(data = breast_cancer_data, columns = breast_cancer.feature_names)
breast_cancer_df

#X_train, X_test, y_train, y_test를 생성
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))

#Decision Tree 사용해 보기
decision_tree = DecisionTreeClassifier(random_state=32)

decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Decision Tree의 결과 값 : ", accuracy)

#Random Forest 사용해 보기
random_forest = RandomForestClassifier(random_state=32)

random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Random Forest의 결과 값 : ", accuracy)

#SVM 사용해 보기
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SVM의 결과 값 : ", accuracy)

#SGD Classifier 사용해 보기
sgd_model = SGDClassifier()

sgd_model.fit(X_train, y_train)

y_pred = sgd_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SGD의 결과 값 : ", accuracy)

#Logistic Regression 사용해 보기
logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression의 결과 값 : ", accuracy)

#학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요?모델의 성능을 평가하는 지표로는 무엇이 좋을까요?
#sklearn.metrics 에서 제공하는 평가지표 중 적절한 것을 선택해 보세요. 선택하신 이유도 설명해 주세요.
print(classification_report(y_test, y_pred))

#Random_Forest가 1.0으로 가장 높은 정확도를 보여주며 다른 모델들 또한 0.9이상의 높은 정확도를 보여준다.
#이는 암환자를 한 명이라도 놓치면 안 되기에 양성을 음성으로 오판하면 안되는 것이 더 중요하여 Recall이 더 중요한 지표로서 선택되여야 한다.

#프로젝트를 진행하면서 알아낸 점
# ==> 이번 프로젝트를 통해 분류에 대해 조금 더 알게 되었고 그 분류를 하는 여러가지 코드와 모델들을 배울 수 있었다.
```
