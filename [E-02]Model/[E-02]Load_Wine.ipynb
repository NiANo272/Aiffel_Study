```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd

#Feature Data 지정하기
wine = load_wine()
wine_data = wine.data

#Label Data 지정하기
wine_label = wine.target

#Target Names 출력해 보기
print(wine.target_names)

#데이터 Describe 해 보기
print(wine.DESCR)

#데이터 프레임 만들기
wine_df = pd.DataFrame(data = wine_data, columns = wine.feature_names)
wine_df

#X_train, X_test, y_train, y_test를 생성
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))

#Decision Tree 사용해 보기
decision_tree = DecisionTreeClassifier(random_state=32)

decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Decision Tree의 결과 값 : ", accuracy)

#Random Forest 사용해 보기
random_forest = RandomForestClassifier(random_state=32)

random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Random Forest의 결과 값 : ", accuracy)

#SVM 사용해 보기
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SVM의 결과 값 : ", accuracy)

#SGD Classifier 사용해 보기
sgd_model = SGDClassifier()

sgd_model.fit(X_train, y_train)

y_pred = sgd_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("SGD의 결과 값 : ", accuracy)

#Logistic Regression 사용해 보기
logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression의 결과 값 : ", accuracy)

#학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요?모델의 성능을 평가하는 지표로는 무엇이 좋을까요?
#sklearn.metrics 에서 제공하는 평가지표 중 적절한 것을 선택해 보세요. 선택하신 이유도 설명해 주세요.
print(classification_report(y_test, y_pred))

#Random_Forest 모델이 1.0으로 높은 수준의 정확도를 보여주었지만 SVM과 SGD는 아주 낮은 성적을 나타내기에 와인 분류에 있어 평면적인 분류를
#하는 SVM이나 최소값을 찾아가는 SGM 모델들은 분류에 적합하지 않다고 볼 수 있다.
#와인 분류를 하며 정상적인 분류를 오판할 지라도 기업의 신뢰도를 위해 올바르지 않은 분류가 일어나지 않는 것이 더 중요하기 때문에 
#Recall지표가 적절한 지표라고 생각한다.
```
